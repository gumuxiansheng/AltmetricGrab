```python
import requests
from requests import ConnectionError

def grab_from_url_content(url):
    headers = {'Accept': '* / *',
               'Accept-Language': 'zh-TW, zh; q=0.9, en-US; q=0.8, en; q=0.7, zh-CN; q=0.6',
               'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/72.0.3610.2 Safari/537.36'
               }
    res = requests.get(url, headers=headers)
    rescontent = res.content

    return rescontent
```

```python
google_cites = []
for doi in df['DI']:
    url = f'https://f.glgoo.top/scholar?hl=en&q={doi}'
    content = grab_from_url_content(url).decode('utf-8')

    anchor = 'Cited by '
    end_anchor = '</a>'

    start_index = content.find(anchor)
    if start_index > 0:
        start_index += len(anchor)
        end_index = content.find(end_anchor, start_index, start_index + 100)
        number = 0
        number_temp = content[start_index: end_index]

        if number_temp is not '':
            number = number_temp
        google_cites.append(number)
    else:
        print(doi)
        google_cites.append(0)
```
